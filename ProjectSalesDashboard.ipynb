```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Atur gaya visualisasi Seaborn
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)  # Ukuran default untuk plot

try:
    df = pd.read_csv('ecommerce_data.csv')
    print("Data berhasil dimuat!")
    print("\nInfo Dataset:")
    df.info()
    print("\n5 Baris Pertama Dataset:")
    print(df.head())
    print("\nDeskripsi Statistik Dataset:")
    print(df.describe())
except FileNotFoundError:
    print("Error: File dataset tidak ditemukan. Pastikan nama file dan lokasinya benar.")
```

```python
# Contoh pra-pemrosesan (sesuaikan dengan dataset Anda!)

# 1. Mengubah nama kolom agar lebih mudah diakses (jika perlu)
# df.columns = df.columns.str.lower().str.replace(' ', '_') # Contoh: jika ada spasi di nama kolom

# 2. Konversi kolom tanggal ke tipe datetime
# Asumsi ada kolom 'OrderDate' atau 'InvoiceDate'
# Cari nama kolom tanggal di dataset Anda (contoh: 'InvoiceDate')
if 'InvoiceDate' in df.columns: # Ganti 'InvoiceDate' dengan nama kolom tanggal Anda
    df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'], errors='coerce')
    print("\nKolom 'InvoiceDate' berhasil diubah ke tipe datetime.")
elif 'OrderDate' in df.columns: # Contoh lain jika nama kolomnya 'OrderDate'
    df['OrderDate'] = pd.to_datetime(df['OrderDate'], errors='coerce')
    df.rename(columns={'OrderDate': 'InvoiceDate'}, inplace=True) # Samakan nama kolom
    print("\nKolom 'OrderDate' berhasil diubah ke tipe datetime dan dinamai ulang menjadi 'InvoiceDate'.")
else:
    print("\nPeringatan: Tidak ditemukan kolom tanggal yang umum seperti 'InvoiceDate' atau 'OrderDate'. Harap sesuaikan!")
    # Jika Anda tahu nama kolom tanggalnya, ubah secara manual di sini:
    # df['NamaKolomTanggalAnda'] = pd.to_datetime(df['NamaKolomTanggalAnda'], errors='coerce')


# Hapus baris dengan nilai tanggal yang tidak valid setelah konversi
df.dropna(subset=['InvoiceDate'], inplace=True)

# 3. Hapus baris dengan kuantitas atau harga negatif/nol (jika ada)
# Asumsi ada kolom 'Quantity' dan 'UnitPrice'
if 'Quantity' in df.columns and 'UnitPrice' in df.columns:
    df = df[(df['Quantity'] > 0) & (df['UnitPrice'] > 0)]
    print("\nBaris dengan kuantitas atau harga <= 0 telah dihapus.")
else:
    print("\nPeringatan: Tidak ditemukan kolom 'Quantity' atau 'UnitPrice'. Harap sesuaikan!")

# 4. Hitung Total Harga (Revenue) untuk setiap baris
if 'Quantity' in df.columns and 'UnitPrice' in df.columns:
    df['TotalPrice'] = df['Quantity'] * df['UnitPrice']
    print("\nKolom 'TotalPrice' telah dibuat.")
else:
    print("\nTidak dapat membuat kolom 'TotalPrice' karena 'Quantity' atau 'UnitPrice' tidak ada.")

# 5. Ekstrak informasi waktu
if 'InvoiceDate' in df.columns:
    df['Year'] = df['InvoiceDate'].dt.year
    df['Month'] = df['InvoiceDate'].dt.month
    df['Week'] = df['InvoiceDate'].dt.isocalendar().week.astype(int)
    df['DayOfWeek'] = df['InvoiceDate'].dt.day_name()
    print("\nKolom 'Year', 'Month', 'Week', dan 'DayOfWeek' telah dibuat.")

print("\nInfo Dataset Setelah Pra-pemrosesan:")
df.info()
print("\n5 Baris Pertama Dataset Setelah Pra-pemrosesan:")
print(df.head())
```

```python
# Agregasi data penjualan bulanan
monthly_sales = df.set_index('InvoiceDate').resample('ME')['TotalPrice'].sum()

# Visualisasi Penjualan Bulanan
plt.figure(figsize=(14, 7))
monthly_sales.plot(kind='line', marker='o', color='skyblue')
plt.title('Tren Penjualan Bulanan (Total Revenue)', fontsize=16)
plt.xlabel('Bulan', fontsize=12)
plt.ylabel('Total Penjualan (Revenue)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Tampilkan data penjualan bulanan
print("\nPenjualan Bulanan:")
print(monthly_sales)
```

```python
# Pastikan InvoiceDate bertipe datetime
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

# Ekstrak tahun dan minggu ISO
df['Year'] = df['InvoiceDate'].dt.isocalendar().year
df['Week'] = df['InvoiceDate'].dt.isocalendar().week

# Agregasi data penjualan mingguan
weekly_sales = df.groupby(['Year', 'Week'])['TotalPrice'].sum().reset_index()

# Buat kolom gabungan untuk label sumbu X
weekly_sales['YearWeek'] = weekly_sales['Year'].astype(str) + '-W' + weekly_sales['Week'].astype(str).str.zfill(2)

# Urutkan berdasarkan Year dan Week
weekly_sales.sort_values(['Year', 'Week'], inplace=True)

# Ambil 20 minggu terakhir
display_weeks = weekly_sales.tail(20)

# Visualisasi penjualan mingguan
plt.figure(figsize=(16, 7))
sns.lineplot(x='YearWeek', y='TotalPrice', data=display_weeks, marker='o', color='lightcoral')
plt.title('Tren Penjualan Mingguan (Total Revenue) - 20 Minggu Terakhir', fontsize=16)
plt.xlabel('Tahun-Minggu', fontsize=12)
plt.ylabel('Total Penjualan (Revenue)', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Tampilkan data mingguan
print("\nPenjualan Mingguan (20 Minggu Terakhir):")
print(display_weeks)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Asumsi ada kolom 'Description' dan 'Quantity'
if 'Description' in df.columns and 'Quantity' in df.columns:
    # Hitung total quantity terjual untuk setiap produk
    top_products_quantity = df.groupby('Description')['Quantity'].sum().nlargest(10).reset_index()

    # Visualisasi Produk Terlaris (Kuantitas)
    plt.figure(figsize=(12, 7))
    sns.barplot(
        data=top_products_quantity,
        x='Quantity',
        y='Description',
        hue='Description',
        palette='viridis',
        dodge=False,
        legend=False  # Mematikan legend sesuai rekomendasi warning
    )
    plt.title('10 Produk Terlaris Berdasarkan Kuantitas', fontsize=16)
    plt.xlabel('Total Kuantitas Terjual', fontsize=12)
    plt.ylabel('Nama Produk', fontsize=12)
    plt.tight_layout()
    plt.show()

    print("\n10 Produk Terlaris Berdasarkan Kuantitas:")
    print(top_products_quantity.set_index('Description')['Quantity'])
else:
    print("\nTidak dapat menganalisis produk terlaris karena kolom tidak ditemukan.")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Asumsi ada kolom 'Description' untuk nama produk dan 'TotalPrice'
if 'Description' in df.columns and 'TotalPrice' in df.columns:
    # Hitung total revenue untuk tiap produk
    top_products_revenue = df.groupby('Description')['TotalPrice'].sum().nlargest(10).reset_index()

    # Visualisasi Produk Terlaris (Revenue)
    plt.figure(figsize=(12, 7))
    sns.barplot(
        data=top_products_revenue,
        x='TotalPrice',
        y='Description',
        hue='Description',         # agar 'palette' bisa digunakan
        palette='magma',
        dodge=False,
        legend=False               # untuk mematikan legenda yang tidak diperlukan
    )
    plt.title('10 Produk Terlaris Berdasarkan Total Penjualan (Revenue)', fontsize=16)
    plt.xlabel('Total Penjualan (Revenue)', fontsize=12)
    plt.ylabel('Nama Produk', fontsize=12)
    plt.tight_layout()
    plt.show()

    print("\n10 Produk Terlaris Berdasarkan Revenue:")
    print(top_products_revenue.set_index('Description')['TotalPrice'])
else:
    print("\nTidak dapat menganalisis produk terlaris berdasarkan revenue karena kolom 'Description' atau 'TotalPrice' tidak ditemukan.")
```

```python
df['TotalPrice'] = df['Quantity'] * df['UnitPrice']
```

```python
df['Category'] = df['Description'].str.split().str[0].fillna('Unknown')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Ambil 10 kategori teratas berdasarkan revenue
top_n = 10
category_sales = (
    df.groupby('Category')['TotalPrice']
    .sum()
    .sort_values(ascending=False)
    .head(top_n)
    .reset_index()
)

# Bar Chart: Total Penjualan per Kategori (Top 10)
plt.figure(figsize=(10, 6))
sns.barplot(
    data=category_sales,
    x='TotalPrice',
    y='Category',
    hue='Category',        # Tambahkan hue
    palette='crest',
    dodge=False,           # Supaya bar tidak terpisah
    legend=False           # Hilangkan legend karena sudah jelas di sumbu Y
)
plt.title(f'Total Penjualan (Revenue) Top {top_n} Kategori', fontsize=14)
plt.xlabel('Total Penjualan (Revenue)', fontsize=12)
plt.ylabel('Kategori Produk', fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.tight_layout()
plt.show()

# Pie Chart: Proporsi per Kategori (Top 10)
plt.figure(figsize=(8, 8))
plt.pie(
    category_sales['TotalPrice'],
    labels=category_sales['Category'],
    autopct='%1.1f%%',
    startangle=140,
    colors=sns.color_palette('pastel')
)
plt.title(f'Proporsi Penjualan (Revenue) Top {top_n} Kategori', fontsize=14)
plt.axis('equal')
plt.tight_layout()
plt.show()
```
